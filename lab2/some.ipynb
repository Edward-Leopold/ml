```python
import pandas as pd

# 1. Прочитайте ваш готовый файл
df = pd.read_csv('lightgbm_final_submission.csv')

# 2. Проверяем, какие столбцы есть
print(f"Текущие столбцы: {list(df.columns)}")

# 3. Переименуйте столбец: если 'Id' (с маленькой d) -> меняем на 'ID'
if 'Id' in df.columns:
    df = df.rename(columns={'Id': 'ID'})
    print("✅ Переименовал 'Id' -> 'ID'")
elif 'ID' in df.columns:
    print("✅ Столбец уже называется 'ID' (всё правильно)")
else:
    print("❌ Нет столбца 'Id' или 'ID' в файле!")
    # Если нет ни того ни другого, создаем 'ID' из индекса
    df.insert(0, 'ID', range(len(df)))
    print("✅ Создал новый столбец 'ID'")

# 4. Проверяем результат
print(f"Исправленные столбцы: {list(df.columns)}")

# 5. Сохраните файл с правильным именем для Kaggle
df.to_csv('submission.csv', index=False)

print("\n✅ Файл 'submission.csv' готов к отправке.")
print("Первые 5 строк:")
print(df.head())
print(f"\nСтолбец называется: {list(df.columns)[0]}")
```
